```python
import numpy as np #importation de la bibliothéque numpy pour effectuer des calculs numéiques
import matplotlib.pyplot as plt # importation du module pyplot de la bibliothéque matplotlib pour la visualisation 
import scipy
import scipy.stats
import time
```

```python
#1_Simuler un échantillon de taille 10000 suivant une loi binomiale B(30, 0.2). 

#n: nombre de fois que l'on onbtient du succé
#p: probabilité d'obtenir du succés
#l: taille de l'échantillon (nombre total d'essai)
#random sous module du module numpy permet de générer des nombres aléatoirement
#binomial fonction de la loi binomiale de paramétre n, p et N (avec N facultatif)

def simu_Binomiale(n,p,N): #définition d'une fonction binomiale
    X = np.random.binomial(n,p,N) #X esera une liste contenant les valeurs de la simulation de la loi 
    return X
print("Notre simulation donne le résultat suivant:",simu_Binomiale(30,0.2,10000))
```

```python
#Tracer l’histogramme de l’échantillon obtenu

simulation = simu_Binomiale(30,0.2,10000) #appele de la fonction simu_Binomiale
plt.rcParams["figure.figsize"] = [16,9] #paramétrage du fenêtre graphique utliser rcParams
plt.hist(simulation, density = True, label="simulation",edgecolor="k") # simulation: valeur d'entrée contenant les arguments de la loi Binomiale,edgecolor permet de distinguer chaque bande de l'histogramme, label = "simulation" place le mot dans la légende,densité: Si Truele premier élément du tuple de retour sera le nombre normalisé pour former une densité de probabilité 
plt.xlabel("Valeur") #axe des abscisses 
plt.ylabel("Probabilité") #axe des ordonnées
plt.title("Loi binomiale") #affiche le titre
plt.draw() #efface les messages au dessus du graphique
plt.savefig("Simulation de la loi Binomiale.png", bbox_inches = "tight")# bbox = "tight" permet de bien proportionner le graphique lors de l'enregistrement
```

```python
#2_Simuler un échantillon de taille 10000 suivant une loi normale N(3, .4)

#n: moyenne de la distribution
#p: écart type de la distribution
#N taille de la distribution
def simu_Normale(n,p,N):
    X = np.random.normal(n,p,N) #échantillon tirés de la distibuton normale paramétrée
    return X
print("Notre simulation donne le résultat suivant:", simu_Normale(3, 0.4,10000))
```

```python
#Tracer la fonction de densité de l’échantillon obtenu. Choisir un intervalle contenant 0 pour domaine de représentation.

#fonction de la densité gaussienne
def densite_gaussienne(x,n,p): 
    return (1/(p*np.sqrt(2*np.pi))*np.exp((-0.5*(x-n)**2)/p**2))  
    #n: moyenne, p: écart-type, pi= 3.1415...                                                            

x = np.linspace(-4,10,200) #permet d’obtenir un tableau 1D allant d’une valeur de départ à une valeur de fin avec un nombre donné d’éléments

plt.plot(x,densite_gaussienne(x,3,0.4) ) #appel de la fonction densité gaussienne dessine la fonction de densité de probabilité de moyenne=3 et d'écart type=0.4, 
plt.grid() #mettre des grilles dans la figure
plt.xlabel("Valeur")
plt.ylabel("Densité")
plt.title("Densité de probabitité de la Loi Normale") #affiche le titre
plt.rcParams["figure.figsize"] = [16,9] #paramétrage du fenêtre graphique utliser rcParams
plt.savefig("Densité de la Loi Normale.png", bbox_inches = "tight")# bbox = "tight" permet de bien proportionner le graphique lors de l'enregistrement
```

```python
#3_Simuler un échantillon de taille 10000 suivant une loi gamma (10, 0.5)

def simu_gamma(n, p, N):
    x = np.random.gamma(n, p, N) #Tire échantillon d'une distribution de Gamma
    #n: forme de distribution de la loi gamma doit être supérieur à 0
    #p: échelle de la distriubution doit être supérieur à 0 par défaut est égal à 1
    #N: taille de l'échantillon
    return x
print("Notre simulation donne le résultat suivant:", simu_gamma(10, 0.5, 10000))
```

```python
#Tracer la fonction de densité de l’échantillon obtenu. Choisir un intervalle contenant 0 pour domaine de représentation

from scipy.stats import gamma    #importe la fonction de densité gamma danss la bibliothéque scipy.stats
gamma_distribution = gamma(loc =10, scale = 10, a = 5)  #la 1iére valeur loc correspond à la moyenne, la deuxiéme valeur scale correspond à l'écart type, la troisiéme valeur a correspond à la paramétre de forme de la densité de gamma
x = np.linspace(-4, 150, 10000)  #linspace est une  fonction numpy qui crée régulièrement espacés structurés comme un tableau numpy; valeur de début = -5, valeur de fin = 200, nombre total de nombre entre la valeur de début et la valeur de fin
#Création de la figure sur un même tracé
fig, ax = plt.subplots(1, 1)
ax.plot(x, gamma_distribution.pdf(x),  lw=4) #pdf: probability density function (fonction de densité de probabilité), lw: line width(nbre de ligne sur la longeur)
plt.rcParams["figure.figsize"] = [15,5] #paramétrage du fenêtre graphique utliser rcParams
plt.title("Densité de probabitité de la Loi Gamma") #affiche le titre
plt.savefig("Densité de la Loi Gamma.png", bbox_inches = "tight")# bbox = "tight" permet de bien proportionner le graphique lors de l'enregistrement
plt.show() #affiche le graphique
```
