```python
#Estimer I2 par une méthode de Monte Carlo avec n = 10000

#n: nombre fois que le répéte l'expérience
#importation de toutes les fonctions du module random: ce dmodule permet de générer des nombres aléatoires
#import de toutes les fonction du module math pour: il fournit des fonctions mathématiques définies par le standard C
import random 
import math
import numpy as np

def estimer(n):  
                       
    cpt=0 #initiale un compteur 
    i=0 #initiale le nbre de fois que l'on répéte l'opération à 0
    while i<n: 
        x=np.random.uniform(low =0, high = 1) #génére aléatoirement un nbre nombbre entre 0 et 1
        y=np.random.uniform(low = 0, high = 1) #génére aléatoirement un nbre nombbre entre 0 et 1
        d = x**2+y**2 #distance qui sépare un point M(x,y) à l'origine 0
        if d <=1: #condition:il faut que la somme des carrés de x et de y soit compris entre 0 et 1
            cpt+=1 #à chaque fois la condition est vérifiée on incrémente le compteur
        i+=1
    print(cpt/n) #si la condition n'est plus vérifiée on sort du boucle for et on affiche le rapport entre le nombre de fois que 
                 #la condition est vérifiée diviser par le nombre total de fois que l'expérience a été répétée: n
```

```python
estimer(10000)
```

```python
#Observer par graphique l’évolution de cette estimation lorsque n varie et vérifier la cohérence avec la
#valeur théorique I2

#Pour répondre répondre réécrivons la fonction précédente
#Ensuite on fera varier la valeur de n  
#Observation 1: n = 10000

import matplotlib.pyplot as plt # importation du module pyplot de la bibliothéque matplotlib pour la visualisation 

n = 10000  #nbre de fois que l'on répéte l'expérience
x = np.random.uniform(low = 0, high = 1, size = [n,1]) #génére aléatoirement un nbre nombbre entre 0 et 1 (expérience répétée n fois)
y = np.random.uniform(low = 0, high = 1, size = [n,1]) #génére aléatoirement un nbre nombbre entre 0 et 1 (expérience répétée n fois)
d = x**2 + y**2 <=1 #distance qui sépare un point M(x,y) à l'origine 0 et elle doit être inférieur à 1
approximation = np.sum(d)/n 
#l'approximation correspond à faire la somme de toutes les distances de n'importe quel point de cordonnées x et y vérifiant 
# la condition d <= 1 divsier par le nombre de fois(n) que cette expérience a été répété.

x_in = x[d] #Un tableau qui contient tous les x générés aléatirement
y_in = y[d] #Un tableau qui contient tous les y générés aléatirement

plt.scatter(x,y, s= 1, label = "nuage de point du carré", color = "g") #nuage de point des x et y du carré
plt.scatter(x_in, y_in,label = "nuage de point du quart de cercle(aire de I2)", color = 'r') #nuage de points des x du quart de cercle
plt.xlabel('Valeurs de x') #axe des abscisses
plt.ylabel('Valeurs de y') #axe des ordonnées
plt.rcParams["figure.figsize"] = [16,9] #paramétrage du fenêtre graphique utliser rcParams
plt.legend(bbox_to_anchor = (1.05, 1), loc = 2, borderaxespad = 0) #légende
plt.title("Graphique évolutive de l'estimation(pour n = 10000)") #affiche le titre
plt.draw() #efface les messages au dessus du graphique
plt.show()
plt.savefig("Graphique évolutive de l'estimation(pour n = 10000).jpeg", bbox_inches = "tight")# bbox = "tight" permet de bien proportionner le graphique lors de l'enregistrement
```
